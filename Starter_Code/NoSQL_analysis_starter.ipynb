### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
count = db['establishments'].count_documents(query)
print("Number of classifications in result: ", count)

# Display the first document in the results using pprint
pprint(db['establishments'].find_one(query))
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
#Convert RatingValue to Integer, so We can use $gte:4
db['establishments'].update_many({}, [
    {"$set": {"RatingValue": {"$toInt": "$RatingValue"}}},
])

# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
#London_query = { "LocalAuthorityName": "London", "RatingValue": { "$gte": 4 } }

#'London' didn't return anything, so return anything with 'London' contained in LocalAuthority
London_query = { "LocalAuthorityName": { "$regex": "london", "$options": "i" }, "RatingValue": { "$gte": 4 } }

# Use count_documents to display the number of documents in the result
london_count = db['establishments'].count_documents(London_query)
print("Number of classifications in result: ", london_count)

# Display the first document in the results using pprint
import pprint
pprint.pprint(db['establishments'].find_one(London_query))

```

```python
# Convert the result to a Pandas DataFrame
results = db['establishments'].find(London_query)
results_list = list(results)
London_df = pd.DataFrame(results_list)

# Display the number of rows in the DataFrame
print("Number of rows in result: ", len(results_list))

# Display the first 10 rows of the DataFrame
London_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

##Find Long and lat of Resteraunt
penang_flavours_doc = establishments.find_one({"BusinessName": "Penang Flavours"})
query_latitude = penang_flavours_doc["geocode"]["latitude"]
query_longitude = penang_flavours_doc["geocode"]["longitude"]

degree_search = 0.01
latitude_range = (query_latitude - degree_search, query_latitude + degree_search)
longitude_range = (query_longitude - degree_search, query_longitude + degree_search)

query = {
    "RatingValue": 5,
    "geocode.latitude": {"$gte": latitude_range[0], "$lte": latitude_range[1]},
    "geocode.longitude": {"$gte": longitude_range[0], "$lte": longitude_range[1]}
}

#create index for hygeine so Mongo is happy when it tries to sory by hygeine score
establishments.create_index([('scores.Hygeine', 1)])

sort  = {'$sort': { 'scores.Hygeine': 1, 'count': 1 }}

limit = 10

results = establishments.find(query).sort([('scores.Hygeine', 1), ('count', 1)]).limit(limit)



# Print the results
print(results)
```

```python
# Convert result to Pandas DataFrame
best_and_nearest_df = pd.DataFrame(results)
best_and_nearest_df.head(5)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': 0}}

# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id': "$LocalAuthorityName", 
                          'count': { '$sum': 1 }}}
# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': { 'count': -1, '_id': 1 }}

# Put the pipeline together
pipeline = [match_query, group_query, sort_values]

# Print the number of documents in the result
results = list(establishments.aggregate(pipeline))
print("Number of classifications in result: ", len(results))

# Print the first 10 results
print(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
Hygeine_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print("Number of rows in result: ", len(results))
# Display the first 10 rows of the DataFrame
Hygeine_df.head(10)
```
